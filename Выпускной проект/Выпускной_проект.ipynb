У нас явно есть минимум 5 больших выбросов вниз в стартовой температуре и 1 большой вниз в итоговой. Избавляться от выбросов будем методом 2 сигм. Сразу напишем функцию, для определения верхнего и нижнего интервалов

```python
# Расчитаем интервал, чтоб избавить от выбросов
def intervals(df, column):
    
    first_quantile = df[column].quantile(0.25)
    third_quantile = df[column].quantile(0.75)
    iqr = third_quantile - first_quantile
    
    interval_low = first_quantile - 2 * iqr
    interval_high = third_quantile + 2 * iqr
        
    return(interval_low, interval_high)
```

Да, результаты очень не очень. Попробуем  библиотеку LGBMR со случайным поиском. Но, поскольку в этой библиотеке нет случайного поиска - возьмем его из sklearn

```python
def algorithm_pipeline(X_train_data, X_test_data, y_train_data, y_test_data, 
                       model, param_grid, cv, scoring_fit):
    
    rs = RandomizedSearchCV(
        estimator=model,
        param_distributions=param_grid, 
        cv=cv, 
        n_jobs=-1,
        scoring=scoring_fit,
        verbose=2
    )
    fitted_model = rs.fit(X_train_data, y_train_data)
    
    
    return fitted_model
```

```python
model= LGBMRegressor()
grid = {
    'max_depth':range(1,16,1),
    'learning_rate':[0.1, 0.5, 1],
    'iterations':range(1,20,1)
}

cv = 3

model = algorithm_pipeline(features_train_scaler, 
                                features_valid_scaler, 
                                target_train, 
                                target_valid, model, 
                                grid, cv, 'neg_mean_absolute_error')
```

```python
depth_LGBM = model.best_params_['max_depth']
iterations_LGBM = model.best_params_['iterations']
learning_rate_LGBM = model.best_params_['learning_rate']

model = LGBMRegressor(depth = depth_LGBM,
                          iterations = iterations_LGBM,
                          learning_rate = learning_rate_LGBM
                          )

model.fit(features_train_scaler, target_train)

mae_LGBMR = mean_absolute_error(target_valid, model.predict(features_valid_scaler))

print('MAE модели LGBMR:',mae_LGBMR)
```

```python



# опеределим зависимые и независимые переменные для обучающей выборки
features= clear_data.drop('finish_temp', axis=1)
target = clear_data['finish_temp']

features_train, features_valid, target_train, target_valid = train_test_split(features, target, 
                                                                                test_size=0.2, random_state=12345)
model= LGBMRegressor()
grid = {
    'max_depth':range(1,16,1),
    'learning_rate':[0.1, 0.5, 1],
    'iterations':range(1,20,1)
}

cv = 3

model = algorithm_pipeline(features_train, 
                                features_valid, 
                                target_train, 
                                target_valid, model, 
                                grid, cv, 'neg_mean_absolute_error')

depth_LGBM = model.best_params_['max_depth']
iterations_LGBM = model.best_params_['iterations']
learning_rate_LGBM = model.best_params_['learning_rate']

model_unscaler= LGBMRegressor(depth = depth_LGBM,
                          iterations = iterations_LGBM,
                          learning_rate = learning_rate_LGBM
                          )

model_unscaler.fit(features_train, target_train)

mae_LGBMR_unscaler = mean_absolute_error(target_valid, model.predict(features_valid))

print('MAE модели LGBMR:',mae_LGBMR_unscaler)
```
