Отлично, тут дублей нет. Но нам необходимо перевести категориальные признаки (тот же пол и страну) в числовой формат. Используем технику ohe. Причем нам следует запомнить, что в случае с географией, если клиент не принадлежит ни Германии, ни Испании, то он из Франции.

```python
data_ohe = pd.get_dummies(data, columns=['gender','geography'], drop_first=True)
display(data_ohe.head(5))
```

Вернемся к пропускам данных. Пометим все отсутствующие как "-1" и посмотрим, с какми данными коррелируют не пропущенные данные. Затем попытаемся восстановить пропущенные. Смотреть будем на коэффицент корреляции Спирмена, а не Пирсона, потому что он более устойчивый к выбросам.

```python
data_ohe['tenure'] = data_ohe['tenure'].fillna(-1)
#print(data_ohe['tenure'].isna().sum())
data_ohe.query('tenure != -1').corr(method='spearman')
```

Столбец tenure прлохо коррелирует с какими бы то ни было столбцами, значит на основании других данных востановить текущие не сильно то и получится. Будем заполнять медианой или средним. Для этого посмотрим на ящик с усами (а лучше на его численное описание) и решим чем лучше заполнять.

```python
data_ohe.query('tenure != -1').boxplot(column=['tenure'])
plt.title('ящик с усами для количества недвижимости во владении')
data_ohe.query('tenure != -1')['tenure'].describe()
```

Отлично. Среднее количество недвижимости в собственности и медианное количество недвижимости в собственности практически совпадают и равны 5. Значит заполним пропуски значением 5.

```python
data_ohe.loc[data_ohe['tenure'] == -1, 'tenure'] = data_ohe.query('tenure != -1')['tenure'].median()

#print(data_ohe.query('tenure ==-1')['tenure'].count())
```

Ну и наконец стандартизируем признаки, приведем их к одному размеру, чтоб модели не принимали, например размер заработной платы, за более важный признак. Для этого разобъем наш датасет на 3 выборки: тренировочную, тестовую и валидационную.


```python
# определим зависимые и независимые переменные
target = data_ohe['exited']
features = data_ohe.drop('exited', axis=1)

# сначала выделим тестовую выборку и тренировочкую
features_train, features_test, target_train, target_test = train_test_split(
    features, target, test_size=0.2, random_state=12345)

# теперь оставшиеся 80% разобъем на тренировочную и валидационную выборку, 
# так, чтоб тренировочная была 60% от датасета, а валидационная 20% от датасета
# другими словами, 75% и 25% от тех 80% что остались

features_train, features_valid, target_train, target_valid = train_test_split(
    features_train, target_train, test_size=0.25, random_state=12345)


# определим все числовие независимые переменные
numeric = ['credit_score', 'age', 'tenure','balance', 'numbrt_of_products', 'estimated_salary']


# обучим стандартизации на тренировочных данных
scaler = StandardScaler()
scaler.fit(features_train[numeric])




# стандартизируем все три выборки
features_train_scaler = features_train.copy() 
features_test_scaler = features_test.copy()
features_valid_scaler = features_valid.copy()

features_train_scaler[numeric] = scaler.transform(features_train[numeric])
features_test_scaler[numeric] = scaler.transform(features_test[numeric])
features_valid_scaler[numeric] = scaler.transform(features_valid[numeric])

#print(features_test_scaler.head())
```

**Вывод**

Загрузили и проверили данные на явные и не явные дубли, а так же  перевели категориальные признаки в цифровые с обходом дамми-ловушки (при этом следует помнить, что если клиент не из Германии и не из Испании, то он из Франции), кроме этого заполнили пропуски данных, разбили данные на 3 выборки и стандартизировали числовые независимые переменные в каждой из них. 

# 2. Исследование задачи

Поскольку наша задача звучит как "Нужно спрогнозировать, уйдёт клиент из банка в ближайшее время или нет", то это означает, что нам необходимо спрогнозировать не конкретное число, а определить, к какому из двух классов ("0" или "1") относится клиент. Поэтому будем использовать модели логистической регрессии, дерева решений и случайного леса. В свою очередь, у модели случайного леса могут быть различные гиперпараметры, такие как глубина (max_depth) и количество деревьев (n_estimators).

Для того чтоб не плодить код сверх необходимого, сделаем функцию, которая будем возвращать параметры

```python
def models(model, features_train, features_valid, target_train, target_valid):
    
    model.fit(features_train,target_train)

    predictions_valid = model.predict(features_valid)
    
    probabilites_valid = model.predict_proba(features_valid)[:,1]

    matrix = confusion_matrix(target_valid, predictions_valid)
    recall = recall_score(target_valid, predictions_valid)
    precision = precision_score(target_valid, predictions_valid)
    f1 = f1_score(target_valid, predictions_valid)
    
    auc_roc = roc_auc_score(target_valid, probabilites_valid)
    
    print(matrix)
    print('полнота:', recall)
    print('точность:', precision)
    print('f1:',f1)
    print('auc-roc:',auc_roc)
    
```

Посмотрим работу алгоритмов на различных моделях.

```python
model_logistic = LogisticRegression(solver='liblinear')
models(model_logistic, features_train_scaler, features_valid_scaler, target_train, target_valid)
```

Модель линейной регрессии достаточно часто правильно предсказывает TN класс, но при этом и часто ошибается, когда встречается с TP. На валидационной выборке она определила всего в 80 из 391 ушедших клиентов. Так дело не пойдет, ведь именно ушедших клиентов мы и пытаемся определить. Хотя у этой модели и хороший показатель auc-roc (что говорит о том, что она лучше чем "случайная" модель), но вот показатель f1 оставляет желать лучшего, а виной тому показатель полноты, который мягко говоря "не достаточно высок".

***Вывод***

Модель логистической регрессии нам не подходит. Выбор остается между моделями дерева решений и моделью случайного леса.

По показателю полноты, модель дерева решений выглядит лучше модели случайного леса. По показателю точности - модель случайного леса лучше модели дерева решений, поэтому будем ориентироваться на производные метрики, такие как f1, который включает в себя и точность и полноту, а так же по показателю auc-roc. По ним модель случайного леса выглядит более убедительной, поэтому с ней и будем продолжать работу.

Так же у нас на лицо явный дисбаланс классов, поэтому модели часто видят негативные классы там где их нет.

# 3. Борьба с дисбалансом

Первым делом попробуем взвесить классы и посмотреть, как модель случайного леса себя поведет.

```python
model_forest = RandomForestClassifier(random_state=12345 , class_weight='balanced')
models(model_forest, features_train_scaler, features_valid_scaler, target_train, target_valid)

```

```python
# создадим функцию для upsamle'инга
def upsample(features, target, repeat):
    #разделим выборку на отрицательные и положительные объекты
    features_zeros = features[target == 0]
    features_ones = features[target == 1]
    target_zeros = target[target == 0]
    target_ones = target[target == 1]
    
    # скопируем несколько раз положительные объекты
    features_upsampled = pd.concat([features_zeros] + [features_ones] * repeat)
    target_upsampled = pd.concat([target_zeros] + [target_ones] * repeat)
    
    # перемешаем строки, так чтоб у нас не было несколько одинаковых блоков, но так чтоб были одинаковые строки
    # по сути, если представить в виде строк, то мы перемешиваем строки таблицы, но не трогаем столбцы
    features_upsampled, target_upsampled = shuffle(
        features_upsampled, target_upsampled, random_state=12345)
    
    return features_upsampled, target_upsampled


# увеличим размер обучающей выборки путем увеличения количества ушедших клиентов
# поскольку изначально доля ушедших клиентов в тренировочной выборке 391/2000= 19.55%, 
# попробуем сделать соотноше ушедших/не ушедших больше и посмотрим, что получится

features_train_scaler_upsampled, target_train_scaler_upsampled = upsample(features_train_scaler, target_train, 5)

models(model_forest, features_train_scaler_upsampled, 
       features_valid_scaler, target_train_scaler_upsampled, target_valid)
```

```python
def improvements(bottom_depth, top_depth, depth_step, bottom_estim, top_estim, estim_step):

    # будем сохранять значение метрики для разных параметров в датафрейм

    results =pd.DataFrame(columns=['depth','estim','recall','precision','roc_auc','f1'])

    # в цикле переберем разные параметры глубины и количества дерерьев в лесу. Для каждого значения посчитаем метрики и запишем таблицу
    for depth in range(bottom_depth, top_depth, depth_step):
        for estim in range(bottom_estim, top_estim, estim_step):
            model = RandomForestClassifier(random_state=12345 , class_weight='balanced', max_depth=depth, n_estimators=estim)
            model.fit(features_train_scaler_upsampled, target_train_scaler_upsampled)

            predictions_valid = model.predict(features_valid_scaler)
            predictions_prob_valid = model.predict_proba(features_valid_scaler)
            
            recall = recall_score(target_valid,predictions_valid)
            precision = precision_score(target_valid,predictions_valid)
            f1 = f1_score(target_valid,predictions_valid)
            auc_roc = roc_auc_score(target_valid,predictions_prob_valid)

            results = results.append({'depth':depth,
                                      'estim':estim,
                                      'recall':recall, 
                                      'precision':precision, 
                                      'roc_auc':auc_roc,
                                      'f1':f1}, 
                                     ignore_index=True)
    return results


bottom_depth = 1
top_depth = 100
depth_step = 10
bottom_estim = 10
top_estim = 100
estim_step = 10

results = improvements(bottom_depth, top_depth, depth_step, bottom_estim, top_estim, estim_step)

results.sort_values(by='f1', ascending = False)
```

```python
display(results.sort_values(by='f1', ascending = False).head(20))
display(results.sort_values(by='f1', ascending = False).head(20).pivot_table(index=['estim','depth'], values=['f1'], aggfunc='count'))
```

```python
results.corr()
```

Отлично, при количестве деревьев в лесу 60 и глубине деревьев  от 31 до 91 мы получили значение f1=0.60 на валидационной выборке. Это конечно хорошо, но попробуем еще улучшить. 

Посмотрим (но уже с шагом 5) леса с 60 деревьями и с глубиной деревьев от 31 до 91, возможно это поможет нам улучшить метрику.

Аналогично можем ориентироваться на значение метрики точности, с которой хорошо однонаправленно коррелирует метрика f1.

```python
%%time 

bottom_depth = 31
top_depth = 91
depth_step = 1
bottom_estim = 60
top_estim = 61
estim_step = 1

results = improvements(bottom_depth, top_depth, depth_step, bottom_estim, top_estim, estim_step)

results.sort_values(by='f1', ascending = False)
```

Как видим, разницы нет. Возьмем модель с количеством деревьев равным 60 и глубиной 31.

Поскольку адекватность модели подтверждается или опровергается путем сравнения со случайной моделью, то можем сравнивать показатель полученный нами с показателем случайной модели, для которой ROC кривая идет из левого нижнего в правый верхний угол, и соответственно площадь под ROC-кривой (та самая AUC-ROC) для случайной модели равно 0.5.
