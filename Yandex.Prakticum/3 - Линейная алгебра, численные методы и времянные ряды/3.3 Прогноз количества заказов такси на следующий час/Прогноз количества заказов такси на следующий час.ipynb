# 3. Обучение

Для обучения нам неплохо было бы разбить нашу подготовленную выборку на тренировочную и тестовую. Сделаем это.

```python
train, test = train_test_split(data, shuffle=False, test_size=0.2)

# проерим корректность разбиения
display(train.index.min(), train.index.max())
display(test.index.min(), test.index.max())
```

Отлично, разбиение прошло корректно. Теперь определим rmse константного прогноза на 1 час вперед. Новое значение x(t) будем прогнозировать предыдущим значением ряда, то есть x(t-1).

```python
pred_previous = test.shift()
pred_previous.iloc[0] = train.iloc[-1].values
rmse_const = mean_squared_error(test,pred_previous)**0.5
print('rmse константного прогноза:',rmse_const)
```

Такой прогноз укладывается в рамки ТЗ. Попробуем что-нибудь другое, вдруг сможем лучше. Для начала создадим признаки: день, месяц, день недели, скользящее среднее длинной в сутки и отстающие значения длинной от часа до суток.

```python
def make_features(data, max_lag, rolling_mean_size):
    data['year'] = data.index.year
    data['month'] = data.index.month
    data['day'] = data.index.day
    data['dayofweek'] = data.index.dayofweek
   
    
    for lag in range(1, max_lag + 1):
        data['lag_{}'.format(lag)] = data['num_orders'].shift(lag)
        
    data['rolling_mean'] = data['num_orders'].shift().rolling(rolling_mean_size).mean()

make_features(data, 24, 24)
```

Разобъем данные снова на тренировочную и тестовую выборки и обучим модель

```python
train, test = train_test_split(data, shuffle=False, test_size=0.2)
train = train.dropna()

features_train = train.drop('num_orders', axis=1)
target_train = train['num_orders']
                     
features_test = test.drop('num_orders', axis=1)
target_test = test['num_orders']                     

model = LinearRegression()
model.fit(features_train,target_train)

pred_test= model.predict(features_test)

print("rmse обучающей выборки:", mean_squared_error(target_test, pred_test)**0.5 )
```

Неплохо получилось rmse относительно константной модели. Посмотрим еще какую-нибудь модель. Пускай это будет модель градиентного бустинга из библиотеки catboost с поиском по сетке и кросс-валидацией. Вот так вот сразу, без объявления войны.

```python
%%time
# По сути это пристрелка, определяем параметры дающие минимальное rmse
# затем попытаемся улучшить подбирая в окресности лучших параметров



cv_dataset = Pool(data=features_train,
                  label=target_train,
                  cat_features=None)




params = {"loss_function": "RMSE"}

scores = cv(cv_dataset,
            params,
            fold_count=3, 
            plot=False,
            as_pandas=True)




model_catboost = CatBoostRegressor()

grid = {'max_depth':range(1,16,4),
        'learning_rate':[0.1, 0.5, 1],
        'iterations':range(1,13,3)
       }

grid_search_result = model_catboost.grid_search(grid, 
                                       X=cv_dataset,
                                       cv=3,
                                       plot=False)
```

```python

depth = grid_search_result['params']['depth']
iterations = grid_search_result['params']['iterations']
learning_rate = grid_search_result['params']['learning_rate']
model = CatBoostRegressor(depth=depth, iterations=iterations, learning_rate=learning_rate)
model.fit(features_train,target_train)
```
