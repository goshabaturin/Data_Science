В этих случаях мы оставляем пропуск и если таких пропусков будет немного, то избавимся от строк с пропусками, если много, то пометим класс как 'unknown'.

И применим на датасете

```python
%%time

# сгруппируем данные по модели, бренду месяцу и году регистрации
# затем возьмем с повторениями из этой группировки только тип авто
# для каждого типа авто найдем самую частую комбинацию модели, бренда, месяца и года регистрации
# сложим тип авто и комбинацию (в виде tuple) в словарь
# при этом по сути получим не словарь, а вектор, где каждый элемент это tuple из 2х компанент: tuple-комбинация и тип авто
mapper = data.groupby(
                    [
                        'Model',
                        'Brand',
                        'RegistrationYear',
                        'RegistrationMonth'
                    ]
                    )['VehicleType'].apply(lambda x: x.mode()).to_dict()

# определим, что весь tuple это key-словаря, а все типы авто ( взятые с повторениями) - это values
mapper ={(key[:-1]):value for key, value in mapper.items()}

def supper_filler(row):
    
    if type(row['VehicleType']) != str:
        
        t = mapper.get((row['Model'],row['Brand'], row['RegistrationYear'], row['RegistrationMonth']),"UNK")
   
    else:
        t = row['VehicleType']
    
    return t

data['VehicleType']=data.apply(supper_filler, axis=1)

display(data['VehicleType'].isna().sum())
```

Касательно года регистрации авто - мы не можем никак восстановить ошибочно забитые данные, поэтому посмотрим на ящик с усами и удалим из датафрейма строки с выбросами в графе "RegistrationYear"

```python
# Расчитаем интервал от Q1-3*IQR до Q3+3*IQR
# все что за границами интервала - выбросы
# если верхняя граница больше 2020, то приравняем верхнюю границу к 2020
def plotting_intervals(column):
    
    first_quantile = data[column].quantile(0.25)
    third_quantile = data[column].quantile(0.75)
    iqr = third_quantile - first_quantile
    
    interval_low = first_quantile - 3 * iqr
    interval_high = third_quantile + 3 * iqr
    
    if interval_high>2020:
        interval_high = 2020
    
    return(interval_low, interval_high)

year_interval_low, year_interval_high = plotting_intervals('RegistrationYear')

# удалим из основного датафрейма строки с выбросами по году регистрации
data = data.loc[data['RegistrationYear'] >= year_interval_low]
data = data.loc[data['RegistrationYear'] <= year_interval_high]

# display(data['RegistrationYear'].min())
# display(data['RegistrationYear'].max())

# снова вглянем на гисторгамму
without_gearbox = data[data['Gearbox'].isna()]
plt.hist(without_gearbox['RegistrationYear'], bins=30)
plt.show()
```

```python
def fill_gearbox(row):
    if type(row['Gearbox']) !=str:
        
        if row['RegistrationYear']<1989:
            gearbox = 'UNK'
        else:
            gearbox = 'cvt'
    else:
        gearbox = row['Gearbox']
        
    return gearbox

data['Gearbox'] = data.apply(fill_gearbox, axis=1)

display(data['Gearbox'].isna().sum())
```

# 2. Обучение моделей

Напомним задачу: заказчик поставил нам задачу разработать модель, которая будет как можно быстрее и как можно точнее предсказывать цену автомобиля, при этом важно еще и минимизировать время обучения. Из задачи сразу вытекает, что нам не следует ориентироваться на время обучение, поэтому  вместо того чтобы разбивать датасет на тренировочный, тестовый и валидационны, будет валидировать модель методом кросс-валидации. Вывод о том, что обучение модели с кросс-валидацией быстрее основан на том, что в классической схеме мы разбиваем фрейм на три 6 частей и 3 из них стандартизируем (features_train, features_valid, features_test), каждую по отдельности. При кросс-валидации мы разбиваем фрейм на 4 части и 2 из них стандартизируем (features_train, features_valid), а 2 вызова функции стандартизации это быстрее чем 3, да и разбить за 1 вызов функции на 4 части тоже быстрее, чем за 2 вызова на 6 частей.

Разобъем данные на features и target и обучим линейную регрессию.

```python
%%time
def train_valid(data):
    # опеределим зависимые и независимые переменные для обучающей выборки
    features = data.drop('Price', axis=1)
    target = data['Price']


    features_train, features_valid, target_train, target_valid = train_test_split(features, target, 
                                                                                  test_size=0.2, random_state=12345)



    #display(features_train.shape)
    #display(target_train.shape)
    #display(features_valid.shape)
    #display(target_valid.shape)

    # определим все числовие независимые переменные
    numeric = features.columns.to_list()


    # обучим стандартизации на тренировочных данных
    scaler = StandardScaler()
    scaler.fit(features_train[numeric])


    # стандартизируем выборки
    features_train_scaler = features_train.copy() 
    features_valid_scaler = features_valid.copy()

    features_train_scaler[numeric] = scaler.transform(features_train[numeric])
    features_valid_scaler[numeric] = scaler.transform(features_valid[numeric])
    
    return features_train_scaler, features_valid_scaler, target_train, target_valid

features_train_scaler, features_valid_scaler, target_train, target_valid = train_valid(data)
```

```python
%%time
def rmse_valid(model, features_valid_scaler, target_valid):
    

    predictions_valid = model.predict(features_valid_scaler)
    
    mse = mean_squared_error(target_valid, predictions_valid) # величина ошибка, без ее направления
    
    
    print('rmse модели: {:.3f}'.format( mse**0.5))
    
    print()
    
    return mse**0.5, predictions_valid



model=LinearRegression()
model.fit(features_train_scaler, target_train)
rmse_linear, predictions_linear = rmse_valid(model, 
                                              features_valid_scaler, 
                                              target_valid)
```

В среднем модель линейной регрессии ошибается на 3.5 тысячи, вроде хорошее предсказание, только вот проблема в том, что средняя стоимость авто в нашем датасете - чуть больше 3 тысяч, а медианная цена и того ниже (да, Series цены положительно скошен).

GrigSearch на моделе случайного леса мы сознательно пропустим, потому как вариативность параметров и скорость их подбора через полное обучение раз за разом не позволят нам достигнуть большой скорости обучения модели, хотя по точности, может и получиться хороший результат.

Начнем с CatBoost и сразу используем встроенные методы кросс-валидации и GridSearch.
```python
%%time
cv_dataset = Pool(data=features_train_scaler,
                  label=target_train,
                  cat_features=None)




params = {"loss_function": "RMSE"}

scores = cv(cv_dataset,
            params,
            fold_count=3, 
            plot=False,
            as_pandas=True)




model_catboost = CatBoostRegressor()

grid = {'max_depth':range(1,16,4),
        'learning_rate':[0.1, 0.5, 1],
        'iterations':range(1,20,5)
       }

randomized_search_result = model_catboost.randomized_search(grid, 
                                       X=cv_dataset,
                                       cv=3,
                                       plot=False)
```

7 минут 25 секунд. Это уже на 16% быстрее. Мы нашли параметры модели, которая дает нам лучшее rmse при использовании библиотеки CatBoost. Теперь обучим модель с этими параметрами и посмотрим на rmse валидационной выборки

```python
%%time
depth_catboost = randomized_search_result['params']['depth']
iterations_catboost = randomized_search_result['params']['iterations']
learning_rate_catboost = randomized_search_result['params']['learning_rate']

model = CatBoostRegressor(depth = depth_catboost,
                          iterations = iterations_catboost,
                          learning_rate = learning_rate_catboost
                          )

model.fit(features_train_scaler, target_train)

rmse_catboost, predictions_catboost = rmse_valid(model, 
                                              features_valid_scaler,  
                                              target_valid)
```

Итого, получается, что на обучение бустинга в библиотеке CatBoost мы потратили 7 минут 31 секунду и получили rmse=1880 на валидационной выборке.

Теперь посмотрим на то, какой скорости и какой точности мы сможем добиться используя модель градиентного бустинга от LightGBM. В этой библиотеке нет вспроенного метода randomizedsearch, поэтому нам придется использовать методов из библиотеки sklearn

```python
def algorithm_pipeline(X_train_data, X_test_data, y_train_data, y_test_data, 
                       model, param_grid, cv, scoring_fit):
    
    rs = RandomizedSearchCV(
        estimator=model,
        param_distributions=param_grid, 
        cv=cv, 
        n_jobs=-1,
        scoring=scoring_fit,
        verbose=2
    )
    fitted_model = rs.fit(X_train_data, y_train_data)
    
    
    return fitted_model
```
